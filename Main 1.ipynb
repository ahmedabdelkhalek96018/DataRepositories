```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
def file(x):
    tree = ET.parse(x)
    root = tree.getroot()
    dfcols = ['id','seo_meta_description','source_id','developer_name']
    df = pd.DataFrame(columns=dfcols)
    def getvalueofnode( root ):
        return root.text if root is not None else None
    for child in root:   
        id = root.attrib.get('id')
        seo_meta_description = root.find('seo_meta_description')
        source_id = root.find('source_id')
        developer_name = root.find('developer_name')
        df = df.append( pd.Series( 
            [id, getvalueofnode(seo_meta_description), getvalueofnode(source_id), getvalueofnode(developer_name)],
            index=dfcols) ,ignore_index=True)
    dup = sum(df.duplicated())
    if dup > 1:
        df.drop_duplicates(inplace = True)
    return df

F1="Explainers/969140360237.explainer.xml"
file(F1)
```
